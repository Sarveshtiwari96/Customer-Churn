```python
Target = df['Churn']
```

```python
df_x = ['gender','SeniorCitizen','Partner', 'Dependents', 'tenure', 'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
 'StreamingTV','StreamingTV', 'Contract', 'PaperlessBilling', 'PaymentMethod', 'MonthlyCharges', 'TotalCharges']
df_dummy = pd.get_dummies(df[df_x])
df_dummy.head()
```

```python
Target = pd.DataFrame(Target)
Target.head()
```

```python
from sklearn.naive_bayes import BernoulliNB,GaussianNB
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier, GradientBoostingClassifier
from sklearn import neighbors, svm, tree

MLA = [
    GaussianNB(),
    BernoulliNB(),
    AdaBoostClassifier(),
    RandomForestClassifier(),
    GradientBoostingClassifier(),
    neighbors.KNeighborsClassifier(),
    svm.SVC(),
    svm.LinearSVC(),
    tree.DecisionTreeClassifier()]
```

```python
import warnings
from sklearn import model_selection
warnings.filterwarnings('ignore')
for i in MLA:
    cv_split = model_selection.ShuffleSplit(n_splits = 10, test_size = .3, train_size = .6, random_state = 0 )
    cv_results = model_selection.cross_validate(i, df_dummy, Target, cv  = cv_split)
    print(cv_results)
    
    print('Train Score for ',str(i).split('(')[0], ' is ', round(cv_results['test_score'].mean(), 3))
#     print('Test Score for ',str(i).split('(')[0], ' is ', round(cv_results['test_score'].mean(), 3))
    print('Fit_time for',str(i).split('(')[0], ' is ', round(cv_results['fit_time'].mean(), 3), 'secs') 
    print('---'*10)
```
